*   отобразите регистрационные номера автомобилей, сгруппированные по сумме штрафов в порядке убывания (мы хотим найти тех, кто заплатил самую большую сумму штрафов);
*   выберите из исходного датафрейма все строки, соответствующие регистрационному номеру с самой большой суммой штрафов (мы хотим проанализировать эти данные более подробно);
*   вывести на экран таблицу, отвечающую на вопрос: есть ли регистрационные номера, которые были установлены на разных моделях автомобилей?



```python
fine_counts = df.reset_index().groupby('CarNumber', as_index=False).agg({'Fines': 'count'}).sort_values('Fines', ascending=False)
fine_counts
```

```python
df.loc[fine_counts.iloc[0, 0]]
```

```python
fines = df.reset_index().groupby('CarNumber', as_index=False).agg({'Fines': 'sum'}).sort_values('Fines', ascending=False)
fines
```

```python
df.loc[fines.iloc[0,0]]
```

```python
diff_models = df.reset_index().groupby('CarNumber', as_index=False).agg({'Model': 'count'}).value_counts('Model').reset_index().sort_values('Model')
diff_models.columns = ['Number_models', 'Number_cars']
diff_models
```
